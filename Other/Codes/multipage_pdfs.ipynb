```python
import numpy
from matplotlib import pyplot as plot
 
# Prepare the data
t = numpy.linspace(-numpy.pi, numpy.pi, 1024)
s = numpy.random.randn(2, 256)
 
#
# Do the plot
#
grid_size = (5, 2)
 
# Plot 1
plot.subplot2grid(grid_size, (0, 0), rowspan=2, colspan=2)
plot.plot(t, numpy.sinc(t), c= '#000000')
 
# Plot 2
plot.subplot2grid(grid_size, (2, 0), rowspan=3, colspan=1)
plot.scatter(s[0], s[1], c= '#000000')
 
# Plot 2
plot.subplot2grid(grid_size, (2, 1), rowspan=3, colspan=1)
plot.plot(numpy.sin(2 * t), numpy.cos(0.5 * t), c= '#000000')
 
# Automagically fits things together
plot.tight_layout()
 
# Done !
plot.show()
```

## Plot a figure with the same certain size and save it in a pdf file  

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.gridspec import GridSpec
import numpy as np

x = np.array([1,2,10]) #np.arange(20)
y = np.array([1,2,3]) #np.arange(10)

fig_width_cm = 21                         # A4 page
fig_height_cm = 29.7
inches_per_cm = 1 / 2.54              # Convert cm to inches
fig_width = fig_width_cm * inches_per_cm # width in inches
fig_height = fig_height_cm * inches_per_cm       # height in inches
fig_size = [fig_width, fig_height]
print(fig_size)

pdf = PdfPages('outfile.pdf')
n_rows = 10  # This is the variable number of subplots
hratios = [1]*n_rows

fig, ax = plt.subplots(n_rows, 1, figsize=fig_size, gridspec_kw={'hspace': 0, 'wspace': 0, 'height_ratios':hratios})
aspect='auto'

if n_rows==1:
    ax.plot(x,y)
    ax.set_aspect(aspect=aspect)
else:
    for j in range(n_rows):
        if j<int(n_rows/2):
            ax[j].plot(x + j, y)
            ax[j].set_aspect(aspect=aspect)
        else: 
            ax[j].axis('off')

pdf.savefig()
pdf.close()
```

## Final plot

```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from matplotlib.gridspec import GridSpec
import numpy as np
import math

x = np.array([1,2,10]) #np.arange(20)
y = np.array([1,2,3]) #np.arange(10)

fig_width_cm = 21                         # A4 page
fig_height_cm = 29.7
inches_per_cm = 1 / 2.54              # Convert cm to inches
fig_width = fig_width_cm * inches_per_cm # width in inches
fig_height = fig_height_cm * inches_per_cm       # height in inches
fig_size = [fig_width, fig_height]
print(fig_size)

pdf = PdfPages('outfile.pdf')
n_plots = 27 # This is the variable number of subplots
n_rows = 8 # This is the maximum number of plots per page
n_cols = 2
hratios = [1]*n_rows

n_plots_left = n_plots
while n_plots_left > 0:
    fig, axs = plt.subplots(n_rows, n_cols, figsize=fig_size, gridspec_kw={'hspace': 0, 'wspace': 0, 'height_ratios':hratios})
    j=0
    for ax in axs.T.reshape(-1):
    #for j in range(n_rows):
        if j < n_plots_left:
            ax.plot(x + j, y)
        else: 
            ax.axis('off')
        j += 1
    pdf.savefig()
    
    n_plots_left -= (j)
    print(n_plots_left)

pdf.close()
```

## Adjust subplots size ratio 

```python
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# create figure
f, ax = plt.subplots(3, 1, figsize=(10,10))

# plot some data
ax[0].plot([1, 2, 3])
ax[1].plot([1, 0, 1])
ax[2].plot([1, 2, 20])

# adjust subplot sizes
gs = GridSpec(3, 1, height_ratios=[5, 5, 10])
for i in range(3):
    ax[i].set_position(gs[i].get_position(f))

plt.show()
```
