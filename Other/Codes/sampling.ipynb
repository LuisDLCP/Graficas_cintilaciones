```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
```

```python
date1 = datetime(2021,6,28,0,0,0)
date1
```

```python
date2 = datetime(2021,6,28,23,59,0)
date2
```

```python
date_today = datetime.now()
date_today
```

```python
days = pd.date_range(start=date1, end=date1+timedelta(1), freq='10T')
days
```

```python
np.random.seed(seed=1111)
data = np.random.randint(1,100, size=len(days))
data
```

```python
df = pd.DataFrame({'Time':days, 'Data': data})
df.head()
```

```python
df.set_index('Time', inplace=True)
df.head()
```

```python
df.sort_index().asfreq('T')
#df.asfreq('T')
```
